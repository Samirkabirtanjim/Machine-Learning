```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

dataset = pd.read_csv('/content/diabetes.csv')

dataset.head()

X = dataset[['Pregnancies', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'Age']]
y = dataset['Glucose']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

regressor = LinearRegression()
regressor.fit(X_train, y_train)

print('Intercept:', regressor.intercept_)
print('Coefficients:', regressor.coef_)

user_input = [float(input(f'Enter the value for {feature}: ')) for feature in X.columns]
user_input = np.array(user_input).reshape(1, -1)

predicted_glucose = regressor.predict(user_input)
print(f'Predicted Glucose Level: {predicted_glucose[0]}')

y_pred = regressor.predict(X_test)
print('Mean Squared Error on Test Set:', metrics.mean_squared_error(y_test, y_pred))
```
