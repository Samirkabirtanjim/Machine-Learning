```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df = df.dropna(subset=['Price'])

X = df[['RAM (GB)', 'Storage Capacity (GB)', 'Battery Capacity (mAh)']]
y = df['Price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
n_neighbors = min(4, len(X_train))

X_train.columns = ['RAM (GB)', 'Storage Capacity (GB)', 'Battery Capacity (mAh)']

knn = KNeighborsRegressor(n_neighbors=n_neighbors)
knn.fit(X_train, y_train)

user_input_ram = float(input("Enter RAM (GB): "))
user_input_storage = float(input("Enter Storage Capacity (GB): "))
user_input_battery = float(input("Enter Battery Capacity (mAh): "))

new_data = [[user_input_ram, user_input_storage, user_input_battery]]
predicted_price = knn.predict(new_data)
print(f'Predicted Mobile Price: {predicted_price[0]}')
```
